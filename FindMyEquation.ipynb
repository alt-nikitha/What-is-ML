```
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

Simple neural network with 1 layer and 1 neuron and 1 value as input

```
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

Next a loss function calculates how far the predicted value is from the original value. An optimiser then tries to minimise this loss. 

```
model.compile(optimizer='sgd', loss='mean_squared_error')
```

We have a set of x values and y values, we need the model to understand the relationship between x and y that defines these pairs.

```
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

The model learns over 500 epochs or training cycles.

```
model.fit(xs, ys, epochs=500)
```

The model is now trained. You can use the model to predict the value of y when x is 10.

```
print(model.predict([10.0]))
```
